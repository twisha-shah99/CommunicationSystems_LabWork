```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Amplitude Modulation

# baseband signal:    m(t) = Am*cos(wm*t)
# carrier signal:     c(t) = Ac*cos(wc*t)
# AM signal:          AM(t) = [Ac + m(t)] * cos(wc*t)
print("Note:")
print("     Amplitude unit: Volt (V)")
print("     Frequency unit: Hertz (Hz)")
print()
A_m = float(input("Enter Baseband Amplitude: "))
f_m = float(input("Enter Baseband Frequency: "))
A_c = float(input("Enter Carrier Amplitude: "))
f_c = float(input("Enter Carrier Frequency: "))

# signal generation

t = np.linspace(0, 1, 1000)

# w_m = 2*pi*f_m, similar for w_c
baseband = A_m*np.cos(2*np.pi*f_m*t)
carrier = A_c*np.cos(2*np.pi*f_c*t)
amplitude_modulated = baseband*carrier

# plotting signals

plt.subplot(3,1,1)
plt.title("Baseband signal (m(t))")
plt.plot(baseband,'g')
plt.xlabel("Time")

plt.subplot(3,1,2)
plt.title("Carrier signal (c(t))")
plt.plot(carrier, 'r')
plt.xlabel("Time")

plt.subplot(3,1,3)
plt.title("AM signal (AM(t))")
plt.plot(amplitude_modulated, color="blue")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.suptitle("Amplitude Modulation")
fig.set_size_inches(16, 11)
```
